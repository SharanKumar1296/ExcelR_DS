***Importing libraries and creating the dataframe***

```python
import pandas as pd
import tensorflow as tf
df = pd.read_csv("gas_turbines.csv")
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
#Finding Null values
df.isna().sum()
```

```python
# split into input (X) and output (Y) variables
X = df.drop("TEY",axis=1)
Y = df["TEY"]
print(X)
print(Y)
```

***Converting the dataset to a single data type to avoid errors***

```python
X_tensor = tf.convert_to_tensor(X, dtype=tf.float64)
Y_tensor = tf.convert_to_tensor(Y, dtype=tf.float64)
```

```python
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
```

```python
model = Sequential()
model.add(Dense(16, input_dim=10,  activation='relu')) #input layer
model.add(Dense(1, activation='sigmoid')) #output layer
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

***Model Fitting***

```python
history = model.fit(X_tensor,Y_tensor, validation_split=0.30, epochs=500, batch_size=200)
```

***Accuracy Score***

```python
scores = model.evaluate(X, Y)
print("%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```

```python
history.history.keys()
```

```python
# summarize history for accuracy
import matplotlib.pyplot as plt
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```
